Загрузите датасет с помощью функции np.genfromtxt()

```python
import numpy as np
url = '/content/iris.data'
species = np.genfromtxt(url,  dtype='str', delimiter=',')
```

Вычислите простейшие статистики: min, max, median, mean, 0.25-percentile,0.75-percentile для каждого из числовых атрибутов в датасете

```python
first = species[:,0]
num_col1=first.astype('float')
second = species[:,1]
num_col2=second.astype('float')
third = species[:,2]
num_col3=third.astype('float')
forth = species[:,3]
num_col4=forth.astype('float')
fifth = species[:,4]
print('Данные для первой колонки: \n' , 'mean: ', num_col1.mean(), ',', 'max: ', num_col1.max(), ',', 'min:', num_col1.min(), ',',
      'median: ',np.median(num_col1),',', 'percentile: ',np.percentile(num_col1, 25),np.percentile(num_col1, 75))
print('Данные для второй колонки: \n' , 'mean: ', num_col2.mean(), ',', 'max: ', num_col2.max(), ',', 'min:', num_col2.min(), ',',
      'median: ',np.median(num_col2),',', 'percentile: ',np.percentile(num_col2, 25),np.percentile(num_col2, 75))
print('Данные для третьей колонки: \n' , 'mean: ', num_col3.mean(), ',', 'max: ', num_col3.max(), ',', 'min:', num_col3.min(), ',',
      'median: ',np.median(num_col3),',', 'percentile: ',np.percentile(num_col3, 25),np.percentile(num_col3, 75))
print('Данные для третьей колонки: \n' , 'mean: ', num_col4.mean(), ',', 'max: ', num_col4.max(), ',', 'min:', num_col4.min(), ',',
      'median: ',np.median(num_col4),',', 'percentile: ',np.percentile(num_col4, 25),np.percentile(num_col4, 75))
```

Через функцию:

```python
atributs = []
for i in range(4):
    atributs.append(np.genfromtxt(url, delimiter=',', dtype='float', usecols=[i]))
for i in range(len(atributs)):
    print(f'Mean {i} = ', np.mean(atributs[i]))
    print(f'Min {i} = ', np.min(atributs[i]))
    print(f'Max {i} = ', np.max(atributs[i]))
    print(f'Median {i} = ', np.median(atributs[i]))
    print(f'Percentile 0.25 для  {i} = ', np.percentile(atributs[i], 25))
    print(f'Percentile 0.75  {i} = ', np.percentile(atributs[i], 75))
    print('\n')
```

Сгруппируйте датасет по категории (последний столбец датасета) и повторите шаги 2 и 3 для групп.

```python
#находим уникальные значения 5 колонки
fifth = species[:,4]
np.unique(fifth)
#Создаем первые 3 сгруппированных датасета по 3м категориям на основе первого числового столбца.
new_1 = []
for i in np.unique(fifth):
    new_1.append(num_col1[fifth==i])
new_1
#Вычисляем простейшие статистики по 3м категориям на основе первого числового столбца
print("Среднее для new_1 = ", np.mean(new_1, axis = 1))
print("Минимум для new_1 = ", np.min(new_1, axis = 1))
print("Максимум для new_1 = ", np.max(new_1, axis = 1))
print("Медиана для new_1 = ", np.median(new_1, axis = 1))
print("Персентиль 0.25 для new_1 = ", np.percentile(new_1, 25, axis = 1))
print("Персентиль 0.75 для new_1 = ", np.percentile(new_1, 75, axis = 1))
#3 Постройте гистограмму для каждого из числовых атрибутов в датасете
plt.hist(new_1)
plt.title("Гистограмма для new_1")
plt.legend([1,2,3])
plt.show()
```

Реализация через цикл:

```python

new_number_atrs = []

for i in range(len(atributs)):
    for group_val in np.unique(fifth): 
        new_number_atrs.append([i, group_val, atributs [i][fifth==group_val]])

for i in range(len(new_number_atrs)):
    print(f'Для числового атрибута {new_number_atrs[i][0]+1}  группы {new_number_atrs[i][1]}')
    print('\n')
    print('*** Выведем статистику - задание 2 ***')
    print(f'Среднее для нового атрибута {i} = ', np.mean(new_number_atrs[i][2]))
    print(f'Минимум для нового атрибута {i} = ', np.min(new_number_atrs[i][2]))
    print(f'Максимум для нового атрибута {i} = ', np.max(new_number_atrs[i][2]))
    print(f'Медиана для нового атрибута {i} = ', np.median(new_number_atrs[i][2]))
    print(f'Персентиль 0.25 для нового атрибута {i} = ', np.percentile(new_number_atrs[i][2], 25))
    print(f'Персентиль 0.75 для нового атрибута {i} = ', np.percentile(new_number_atrs[i][2], 75))
    print('\n')
    print('*** Выведем гистограмму - задание 3 ***')
    plt.hist(new_number_atrs[i][2])
    plt.title(f'Гистограмма для нового атрибута {i}')
    plt.show()
```
